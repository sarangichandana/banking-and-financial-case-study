### Checking for columns having null values and how to treat them

```python
#checking for columns having greater than 50% null values
emptycol=df.isnull().sum()/len(df)*100
emptycol=emptycol[emptycol.values>50]
emptycol
```

```python
len(emptycol)
```


#### Insight & Tratment:
 - Columns above 50% nulls -  Above we can see there are 41 columns having above 50% null values So we will not consider these columns in our analysis. **Treatment** - Drop the columns
 - Columns below 50% nulls - below 50% columns we will impute mean or median or mode
 

```python
#removing all columns having more than 50% null values
emptycol=list(emptycol[emptycol.values>=50].index)
df.drop(labels=emptycol,axis=1,inplace= True)
len(emptycol)
```

```python
# Binning   

def amount_annuity(x):
    if x <=25000:
        return "low"
    elif x>25000 and x<=50000:
        return "medium"
    elif x>50000 and x<=100000:
        return "high"
    else:
        return "very high"

    
df['AMT_ANUITY'] = df['AMT_ANNUITY'].apply(lambda x: amount_annuity(x))
```
