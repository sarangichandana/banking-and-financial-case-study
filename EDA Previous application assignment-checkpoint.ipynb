## Merging dataset by SK_ID_CURR

```python
final = df.merge(df1, left_on='SK_ID_CURR', right_on='SK_ID_CURR', how='inner')
final.head()
```

```python
final.shape
```

```python
# Separate defaulter & non-defaulter
final_0 = final[final['TARGET'] == 0]
final_1 = final[final['TARGET'] == 1]
```

## Univariate Analysis

```python
categ_col10 = ['CODE_GENDER','NAME_EDUCATION_TYPE','NAME_CONTRACT_TYPE_x','NAME_CONTRACT_TYPE_y','NAME_TYPE_SUITE_x','NAME_TYPE_SUITE_y','NAME_INCOME_TYPE']
```

```python
plt.figure(figsize=(16,25))
for ind,col in enumerate(categ_col10[:-1]):
    
    plt.subplot(len(categ_col10)//2,2,ind+1)
    sns.countplot(x = col, data = final, hue = "NAME_CONTRACT_STATUS")
    plt.title(f"Univariate Analysis : {col}\n",weight="bold",fontsize=11)
    plt.xticks(rotation=90)
    plt.xlabel(" ")
    plt.ylabel("Count")
    plt.legend(fontsize=10,loc='upper right')

    plt.tight_layout()
    plt.show
```

Observation :

- Maximum no.of loans are approved in case of female in CODE_GENDER than males.
- Secondary Special sector has got more loans approved by the bank in comparison to others in NAME_EDUCATION_TYPE.
- Unaccompanied has the highest number in both NAME CONTRACT TYPE Previous and NAME CONTRACT TYPE Current.
- Earlier the bank had offered cash, revolving loans. But in NAME_CONTRACT_TYPE_y bank has given consumer loans in addition to the existing loans which is got approved the most. 

```python
plt.figure(figsize=(15,20))
sns.countplot(x = 'ORGANIZATION_TYPE', data = final, hue = "NAME_CONTRACT_STATUS")
plt.xticks(rotation=90,weight='bold')
plt.show()
plt.tight_layout()
```

```python
# NUMERICAL ANALYSIS 
num_col = ['AMT_ANNUITY_x','AMT_ANNUITY_y','AMT_CREDIT_x','AMT_CREDIT_y','AMT_GOODS_PRICE_x','AMT_GOODS_PRICE_y']
```

```python
plt.figure(figsize=(22,25))
for ind,col in (enumerate(num_col)):
    plt.subplot(len(num_col)//2,2,ind+1)
    sns.distplot(final.loc[final.NAME_CONTRACT_STATUS=='Unused offer',:][col].dropna(),hist=False,label='Unused offer',color = "darkblue")
    sns.distplot(final.loc[final.NAME_CONTRACT_STATUS=='Canceled',:][col].dropna(),hist=False,label='Cancelled',color="yellow")
    sns.distplot(final.loc[final.NAME_CONTRACT_STATUS=='Approved',:][col].dropna(),hist=False,label='Approved', color="b")
    sns.distplot(final.loc[final.NAME_CONTRACT_STATUS=='Refused',:][col].dropna(),hist=False,label='Refused',color="red")
    plt.title(f"Univariate Analysis : {col}\n",weight="bold",fontsize=11)
    plt.xlabel(" ")
    plt.legend(loc="upper right")
    plt.tight_layout()
```
